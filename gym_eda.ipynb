```python
plt.figure(figsize=(10, 6))
df.groupby('Age')['Calories_Burned'].sum().plot(kind='line', color='deepskyblue', linewidth=3, marker='o', markersize=7, linestyle='-', markerfacecolor='red')
plt.title('Total Calories Burned Across Different Age Groups', fontsize=18, fontweight='bold', color='darkblue')
plt.xlabel('Age', fontsize=14, fontweight='bold', color='darkgray')
plt.ylabel('Calories Burned', fontsize=14, fontweight='bold', color='darkgray')
plt.grid(True, which='both', linestyle='--', linewidth=0.7, alpha=0.5)
plt.xticks(fontsize=12, rotation=45, color='darkgray')
plt.yticks(fontsize=12, color='darkgray')
plt.tight_layout()
plt.show()
```

```python
experience_levels = df['Experience_Level'].unique()
session_durations_by_experience = [df.groupby('Experience_Level')['Session_Duration (hours)'].get_group(level).values for level in experience_levels]

plt.figure(figsize=(10, 6))
plt.boxplot(session_durations_by_experience, patch_artist=True,
            boxprops=dict(facecolor='#1f77b4', color='black'),
            whiskerprops=dict(color='black', linewidth=1.5),
            flierprops=dict(markerfacecolor='red', marker='o', markersize=8, linestyle='none'),
            medianprops=dict(color='yellow', linewidth=2))
plt.title('Session Duration by Experience Level', fontsize=18, fontweight='bold', color='darkblue')
plt.xlabel('Experience Level', fontsize=14, fontweight='bold', color='darkgray')
plt.ylabel('Session Duration (hours)', fontsize=14, fontweight='bold', color='darkgray')
plt.xticks(np.arange(1, len(experience_levels) + 1), experience_levels, fontsize=12, color='darkgray')
plt.grid(True, axis='y', linestyle='--', linewidth=0.7, alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
payment_counts = df['Experience_Level'].value_counts()
colors = ['#4B8BBE', '#E5B95B', '#A55E79', '#8B8B8B', '#4CAF50']
plt.figure(figsize=(6, 6))
payment_counts.plot(kind='pie',
                    autopct='%1.1f%%',
                    startangle=140,
                    colors=colors,
                    explode=[0.1] * len(payment_counts),
                    shadow=True)


plt.title('Experience Level Distribution Among Participants', fontsize=16, fontweight='bold')
plt.ylabel('')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
labels = ['Male', 'Female']
sizes = [52.52, 47.48]
colors = ['#1f77b4', '#ff69b4']
explode = (0.1, 0)
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.2f%%', shadow=True, startangle=140,
        textprops={'fontsize': 14})
plt.title('Gender Representation in the Sample Population', fontsize=16, fontweight='bold')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
mean_fat_percentage = df.groupby('Workout_Type')['Fat_Percentage'].mean().sort_values()
plt.figure(figsize=(10, 6))
mean_fat_percentage.plot(kind='bar', color='purple', edgecolor='black')
plt.title('Average Fat Percentage by Workout Type', fontsize=14, fontweight='bold')
plt.ylabel('Average Fat Percentage', fontsize=12)
plt.xlabel('Workout Type', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
Avg_BPM_and_weight = df.groupby(['Weight (kg)', 'Water_Intake (liters)'])['Avg_BPM'].sum().unstack()
Avg_BPM_and_weight.plot(
    kind='area',
    stacked=True,
    figsize=(12, 6),
    alpha=0.7,
    title='Average BPM vs Water Intake Across Different Weights'
)
plt.ylabel('Average BPM', fontsize=12)
plt.xlabel('Weight (kg)', fontsize=12)
plt.title('Average BPM vs Water Intake Across Different Weights', fontsize=14, fontweight='bold')
plt.grid(alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df.groupby('Weight (kg)')['BMI'].sum().plot(kind='line', color='blue', title='Weight impact on BMI')
plt.ylabel('Weight (kg))')
plt.xlabel('BMI')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Session_Duration (hours)'], df['Calories_Burned'],
            color='blue', edgecolor='w', s=100, alpha=0.7)
plt.title('Relationship Between Session Duration and Calories Burned', fontsize=16, fontweight='bold')
plt.xlabel('Session_Duration (hours)', fontsize=14)
plt.ylabel('Calories_Burned', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
m, b = np.polyfit(df['Session_Duration (hours)'], df['Calories_Burned'], 1)
plt.plot(df['Session_Duration (hours)'], m * df['Session_Duration (hours)'] + b, color='red', linewidth=2, linestyle='--')
plt.tight_layout()
plt.show()
```

```python
mean_workout_frequency = df.groupby('Experience_Level')['Workout_Frequency (days/week)'].mean()
plt.figure(figsize=(12, 6))
mean_workout_frequency.plot(kind='bar', color='skyblue')
plt.title('Average Workout Frequency by Experience Level', fontsize=16, fontweight='bold')
plt.xlabel('Experience Level', fontsize=14)
plt.ylabel('Average Workout Frequency (days/week)', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
Calories_Burned = df['Calories_Burned']
Fat_Percentage = df['Fat_Percentage']
plt.figure(figsize=(10, 6))
scatter = plt.scatter(Calories_Burned,
                      Fat_Percentage,

                      alpha=0.6,
                      c=Fat_Percentage,
                      cmap='viridis',
                      edgecolors='w',
                      linewidth=2)
cbar = plt.colorbar(scatter)
cbar.set_label('Body Fat Percentage (%)')
plt.title('Relationship Between Calories Burned and Body Fat Percentage', fontsize=16)
plt.xlabel('Calories_Burned', fontsize=14)
plt.ylabel('Body Fat_Percentage', fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()
```
